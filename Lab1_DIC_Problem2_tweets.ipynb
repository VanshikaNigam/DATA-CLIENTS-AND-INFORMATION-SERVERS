```r
#http://stackoverflow.com/questions/18115550/how-to-combine-two-or-more-columns-in-a-dataframe-into-a-new-column-with-a-new-n
lat_long$combine <- paste(lat_long$lon,lat_long$lat)
#http://stackoverflow.com/questions/1923273/counting-the-number-of-elements-with-the-values-of-x-in-a-vector
store_freq<-as.data.frame(table(lat_long$combine))
print(store_freq)
```

```r
#http://stackoverflow.com/questions/6709151/how-do-i-combine-two-data-frames-based-on-two-columns
df_merge_final=merge(lat_long, store_freq, by.x=c("combine"), by.y=c("Var1"))
show(df_merge_final)
```

```r
#plotting the map for superbowl
map<- get_googlemap(location="United States",zoom=4,maptype="roadmap",color="color")
map_usa<-ggmap(map)
map_usa<-map_usa+geom_point(aes(x = lon, y = lat), data = df_merge_final, size = df_merge_final$Freq/3, colour ="purple")
print(map_usa)
```
